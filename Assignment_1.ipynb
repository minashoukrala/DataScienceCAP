```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Gender": ["male", "male", "female"],
        "Height": [151, 143, 190]
    }
)

df
```

```python
# display the second person

df.loc[1]
```

```python
# display all ages

df["Age"]
```

```python
# display maximum age

df["Age"].max()
```

```python
# get the index of the oldest person

df['Age'].idxmax()
```

```python
# filter by females only

females_df = df[df['Gender'] == 'female']
females_df
```

```python
# add a new column "Profession"

df["Profession"] = ["Engineer", "Doctor", "Accountant"]
df
```

```python
# the max age per Gender

df[["Gender", "Age"]].groupby("Gender").max()
```
